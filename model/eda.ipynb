```python
# =============================
# 1. Import library
# =============================
import pandas as pd

# =============================
# 2. Load dataset
# =============================
df = pd.read_csv("data/ObesityDataSet_raw_and_data_sinthetic.csv")

# Lihat 5 baris pertama
print("=== Head ===")
df.head()
```

```python
# =============================
# 3. Informasi umum dataset
# =============================
print("=== Info Data ===")
df.info()
```

```python
# =============================
# 4. Statistik deskriptif
# =============================
print("=== Deskripsi Statistik (Numerik) ===")
print(df.describe(), "\n")

print("=== Deskripsi Statistik (Kategorikal) ===")
print(df.describe(include="object"), "\n")
```

```python
# =============================
# 5. Cek missing values
# =============================
print("=== Missing Values ===")
print(df.isnull().sum(), "\n")

# =============================
# 6. Cek jumlah nilai unik per kolom
# =============================
print("=== Unique Values per Column ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
print("\n")
```

```python

# =============================
# 7. Distribusi kelas target (jika ada kolom label)
# =============================
target_col = "NObeyesdad"   # ganti jika nama kolom berbeda

if target_col in df.columns:
    print("=== Distribusi Target ===")
    print(df[target_col].value_counts())
    print("\n")

# =============================
# 8. Cek korelasi untuk fitur numerik
# =============================
numeric_df = df.select_dtypes(include=['int64', 'float64'])
print("=== Korelasi Numerik ===")
print(numeric_df.corr(), "\n")

# =============================
# 9. Cek outliers sederhana (IQR)
# =============================
print("=== Outlier Detection (IQR Method) ===")
for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((numeric_df[col] < (Q1 - 1.5 * IQR)) | (numeric_df[col] > (Q3 + 1.5 * IQR))).sum()
    print(f"{col}: {outliers} outliers")
```
